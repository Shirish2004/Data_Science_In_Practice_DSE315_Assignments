## Machine Learning 

### There are numerous types of classifier that can be used for this classification task. We cannot test each of the classifer hence we will be using a library called the "Lazy Predict" that tests all the basic classifiers on its basic parameters to perform the classification task and ranks them according to their performance. From there we can select the best classifer and Fine tune Hyper parameters to make it more robust

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
X = df_final[columns]#feature variables
y= df_final.INDREC#target variable
display(X,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=123)

clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)

print(models)
```

### From above we can state that the ExtraTreesClassifier hasa performed the best and has also taken very less time. We will be using for the model building 

```python
from sklearn.ensemble import ExtraTreesClassifier

classifier=ExtraTreesClassifier(n_estimators=200,criterion="entropy",bootstrap=True,random_state=42)

classifier.fit(X_train,y_train)


```

```python
y_hat=classifier.predict(X_test)
y_hat
```

```python
#getting classification report 
from sklearn.metrics import classification_report, confusion_matrix,roc_curve,roc_auc_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler

#getting classification report 
classification_report=classification_report(y_test,y_hat)
print(classification_report)
```

```python
#getting confusion matrix 
matrix=confusion_matrix(y_test,y_hat)
print(matrix)
```

```python
#getting roc_auc_score

roc_auc_score=roc_auc_score(y_test,y_hat)

print("The Roc_Auc_Score of the classifier is: ",roc_auc_score)
```

##### A model is considered deployable when it has an Roc_Auc_Score more than 92 thus, we can say that the ExtraTreesClassifier is a good model

```python
#plotting confusion matrix
plt.style.use('seaborn')
from sklearn.metrics import plot_confusion_matrix
cm=plot_confusion_matrix(classifier,X_test,np.array(y_test).reshape(-1,1))
cm
plt.savefig("confussion matrix",dpi=300,bbox_inches="tight")
plt.show()
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_hat).ravel()
print(f'True Positives: {tp}')
print(f'False Positives: {fp}')
print(f'True Negatives: {tn}')
print(f'False Negatives: {fn}')
```

```python
probas = classifier.predict_proba(X_test)[:, 1]
probas#predicted probabilities for each class in X_test
```

```python
def get_preds(threshold, probabilities):
    return [1 if prob > threshold else 0 for prob in probabilities]
roc_values = []
for thresh in np.linspace(0, 1, 100):
    preds = get_preds(thresh, probas)
    tn, fp, fn, tp = confusion_matrix(y_test, y_hat).ravel()
    tpr = tp/(tp+fn)
    fpr = fp/(fp+tn)
    roc_values.append([tpr, fpr])
# tpr_values, fpr_values = zip(*roc_values)
fpr, tpr, _ = roc_curve(y_test,  probas)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
font_2 = {'family': 'sans-serif',
        'color':  'white',
        'weight': 'normal',
        'size': 20,
        }
ax.plot(fpr, tpr,linewidth=4,linestyle="dashdot",color="r")
ax.plot(np.linspace(0, 1, 100),
         np.linspace(0, 1, 100),
         label='baseline',
         linestyle='--')
plt.title('Receiver Operating Characteristic Curve', fontdict=font_2)

plt.ylabel('TPR', fontsize=16)
plt.xlabel('FPR', fontsize=16)
plt.legend(fontsize=12)

ax.text(0.4,0.2,f"AUC Score {roc_auc_score}",fontdict=font_2,bbox={'facecolor': 'black','alpha':0.8, 'pad':15})
plt.grid(True)
plt.savefig("ROC-AUC-Curve",dpi=300,bbox_inches="tight")
plt.show()
```
